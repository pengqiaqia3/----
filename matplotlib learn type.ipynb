### 风格改变

```python
import matplotlib.pyplot as plt
import numpy as np
#将 matplotlib 生成的图表直接嵌入到 Jupyter Notebook 的输出单元格中，便于查看和编辑
%matplotlib inline  
```

```python
plt.style.available  # 显示所有可以用的风格
```

```python
x = np.linspace(start=-10,stop=10)
y = np.sin(x)
plt.plot(x,y,linewidth = 3.0)
plt.style.use('dark_background')
```

```python
plt.plot(x,y)
plt.style.use('bmh')
```
