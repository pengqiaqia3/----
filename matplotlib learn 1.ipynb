```python
import matplotlib.pyplot as plt
import numpy as np
#将 matplotlib 生成的图表直接嵌入到 Jupyter Notebook 的输出单元格中，便于查看和编辑
%matplotlib inline  
```

### 最基本的的图

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.rc('font',family='YouYuan')  # 设置字体用来显示中文
plt.xlabel('x-轴',fontsize=16)
plt.ylabel('y-轴')
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'--')  # 自定义线条格式
plt.rc('font',family='YouYuan')  # 设置字体用来显示中文
plt.xlabel('x-轴',fontsize=16)
plt.ylabel('y-轴',fontsize=16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.',color='r')  # 自定义线条格式
plt.rc('font',family='YouYuan')  # 设置字体用来显示中文
plt.xlabel('x-轴',fontsize=16)
plt.ylabel('y-轴',fontsize=16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'r-.')  # 自定义线条格式
plt.rc('font',family='YouYuan')  # 设置字体用来显示中文
plt.xlabel('x-轴',fontsize=16)
plt.ylabel('y-轴',fontsize=16)
```

### 绘制多条线

```python
qin_numpy = np.arange(0,10,0.5)
plt.plot(qin_numpy,qin_numpy,'r--')
plt.plot(qin_numpy,qin_numpy**2,'bs')
plt.plot(qin_numpy,qin_numpy**3,'g-.')
```

```python
plt.plot(qin_numpy,qin_numpy,'r--',
         qin_numpy,qin_numpy**2,'bs',
         qin_numpy,qin_numpy**3,'g-.')
```

### 指定线条宽度

```python
x = np.linspace(start=-10,stop=10)
y = np.sin(x)
plt.plot(x,y,linewidth = 3.0)
```

### 标记点样式

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
```

### 透明度

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth=2.0,alpha=0.2)  #设置线条宽度 透明程度
```

### 子图

```python
plt.subplot(211)  #211 表示图片是2行1列的  第1个子图
plt.plot(x,y,color='r')

plt.subplot(212)  #212 表示图片是2行1列的  第2个子图
plt.plot(x,y,color='b')
```

```python
plt.subplot(121)  #211 表示图片是1行2列的  第1个子图
plt.plot(x,y,color='r')

plt.subplot(122)  #212 表示图片是1行2列的  第2个子图
plt.plot(x,y,color='b')
```

```python
plt.subplot(321)  #211 表示图片是3行2列的  第1个子图
plt.plot(x,y,color='r')

plt.subplot(324)  #212 表示图片是1行2列的  第4个子图
plt.plot(x,y,color='b')
```

### 加入图片的注释 

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('sin(x)函数')
plt.text(0,0,'原点')
plt.grid(True)   # 加入方格
# xy 为标记点位置  xytext 为标签位置  arrowprops 为箭头设置
plt.annotate(text='函数最大值',xy=(np.pi/2,np.sin(np.pi/2)),xytext=(-4,1),arrowprops=dict(facecolor='red',shrink=0.05,headlength=20,headwidth=20))   # 箭头标记设置
```
